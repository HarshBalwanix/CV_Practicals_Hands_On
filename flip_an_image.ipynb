```python
import cv2
from matplotlib import pyplot as plt
```

***Reading the image*** 

```python
img =cv2.imread('opencv.png')
name='OpenCv python Image'
plt.title(name)
plt.imshow(img)
```

***Flipping the image horizontally***

```python
horizontal_flip=cv2.flip(img,1) # 1 denotes around Y axis
name='OpenCV Python flip image Horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

***Flipping the image Vertically ****


```python
vertical_flip=cv2.flip(img,0) # 0 denotes around X axis 
name='OpenCv python Flip image vertically'
plt.title(name)
plt.imshow(vertical_flip)
```

***Flipping the image Vertically and Horizontally ****

```python
vertical_horizontal_flip=cv2.flip(img,-1) # fliped in both the axes
name='OpenCV Python Flip Image in Both Axes'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```
